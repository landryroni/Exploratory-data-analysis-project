### Top 20 teams that have highest number of international matches.

To visualize the top 20 teams we highest number of internationales matches played we use barplot to plot it. Here we to use the`football_df` dataframe which is the main dataframe, then need to get total `home_team` played matches plus total `away_team` played matches then `sum` it.

```python
def total_games(football_df):
  matches_df = football_df
  total_games_matches = matches_df.agg({'home_team': 'value_counts', 'away_team': 'value_counts'}).fillna(0)
  total_games_matches["total"] = total_games_matches.eval("home_team + away_team")
  total_games_matches = total_games_matches.sort_values("total",ascending=False)
  return total_games_matches 
top_total_matches= total_games(football_df)
top20_team_total_matches = top_total_matches.sort_values(by="total",ascending=False).head(20)
top20_team_total_matches = top20_team_total_matches.reset_index().rename(columns={"index":"country"})
top20_team_total_matches
```

```python
plt.figure(figsize=(15,10))
g = sns.barplot(x='total', y="country", data = top20_team_total_matches)
def show_values_on_bars(axs, h_v="v", space=0.4):
    def _show_on_single_plot(ax):
        if h_v == "v":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() / 2
                _y = p.get_y() + p.get_height()
                value = int(p.get_height())
                ax.text(_x, _y, value, ha="center") 
        elif h_v == "h":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() + float(space)
                _y = p.get_y() + p.get_height()
                value = int(p.get_width())
                ax.text(_x, _y, value, ha="left")

    if isinstance(axs, np.ndarray):
        for idx, ax in np.ndenumerate(axs):
            _show_on_single_plot(ax)
    else:
        _show_on_single_plot(axs)
show_values_on_bars(g, h_v="h", space=0.4)
plt.title("Top 20 contries team with highest total matches played including friendly macthes")
plt.ylabel("Team")
plt.xlabel("Total number of Matches")
plt.show()
```

### Number of FIFA major tournaments matches played by each country national team

```python
def total_fifa_games(football_df):
  matches_df = football_df[(football_df.tournament).isin(fifa_major_tournaments)]
  total_fifa_games_matches = matches_df.agg({'home_team': 'value_counts', 'away_team': 'value_counts'}).fillna(0)
  total_fifa_games_matches["total"] = total_fifa_games_matches.eval("home_team + away_team")
  total_fifa_games_matches = total_fifa_games_matches.sort_values("total",ascending=False)
  return total_fifa_games_matches 
```

### The evolution of number teams participating in each FIFA major tounament. 

This part is just to show the history of how Fifa and other continent football federation increased the number of team gradually. Many reason have been given, like to give more chances to several countries to participate and make the competition more challenging and attractive.

```python
def total_team_per_cup(football_df):
  games_per_tournaments_df = football_df[["date","home_team","away_team","tournament"]]
  games_per_tournaments_df = games_per_tournaments_df.loc[(games_per_tournaments_df.tournament).isin(fifa_major_tournaments)]
  games_per_tournaments_df["date"] = games_per_tournaments_df.date.dt.year
  games_per_tournaments_df =  games_per_tournaments_df.groupby(["date","tournament"])["home_team","away_team"].agg({'home_team': 'value_counts', 'away_team': 'value_counts'}).fillna(0)
  games_per_tournaments_df = games_per_tournaments_df.reset_index().rename(columns={"level_0":"date","level_1":"tournament","level_2":"team"}).drop(columns=["home_team","away_team"])
  games_per_tournaments_df = games_per_tournaments_df.set_index("date")
  games_per_tournaments_df = games_per_tournaments_df.groupby(["date","tournament"]).count()
  # games_per_tournaments_df = pd.Index(games_per_tournaments_df.index).value_counts() 
  games_per_tournaments_df = games_per_tournaments_df.reset_index()
  return games_per_tournaments_df#.set_index("date")
```

```python
team_per_competition = total_team_per_cup(football_df)
team_per_competition
```

```python
px.bar(team_per_competition,x="date",y="team",color="tournament",title="Evolution of Number of teams participating in Fifa major tournament in each edition by Continent",labels={"team":"Number of Team","date":"Date of Edition"})
```

### Statistic of top 20 teams(Win, draw, loss) 

Here we analyse the top 20 Teams statistic by looking for their `home` played game and `away` played game result. We use the `football_df` dataframe.

```python
teams = football_df.agg({"home_team":"value_counts","away_team":"value_counts"})
teams = teams.reset_index()
teams = list(teams["index"])
# teams
```

```python
# getting the total home and away win,loss,draw matches result
def function_ha(new_df_wins,teams):
  win_list_h,loss_list_h,draw_list_h,team_list_h = [],[],[],[]
  win_list_a,loss_list_a,draw_list_a,team_list_a = [],[],[],[]
  
  for i in range(len(new_df_wins)):
      wins_h = 0
      losses_h = 0
      draw_h = 0
      if new_df_wins['home_team'][i] in teams:
          if new_df_wins['home_score'][i]>new_df_wins['away_score'][i]:
              wins_h+=1
          elif new_df_wins['home_score'][i]<new_df_wins['away_score'][i]:
              losses_h+=1
          else:
              draw_h+=1

      team_list_h.append(new_df_wins['home_team'][i])      
      win_list_h.append(wins_h)
      loss_list_h.append(losses_h)
      draw_list_h.append(draw_h)
  for i in range(len(new_df_wins)):
      wins_a = 0
      losses_a = 0
      draw_a = 0
      if  new_df_wins['away_team'][i] in teams:
          if new_df_wins['home_score'][i]<new_df_wins['away_score'][i]:
              wins_a+=1
          elif new_df_wins['home_score'][i]>new_df_wins['away_score'][i]:
              losses_a+=1
          else:
              draw_a +=1
      team_list_a.append(new_df_wins['away_team'][i])      
      win_list_a.append(wins_a)
      loss_list_a.append(losses_a)
      draw_list_a.append(draw_a)

  return {"team_names_h":team_list_h,
        "win_h":win_list_h,
        "loss_h":loss_list_h,
        "draw_h":draw_list_h}, {"team_names_a":team_list_a,
        "win_a":win_list_a,
        "loss_a":loss_list_a,
        "draw_a":draw_list_a}
```

```python
#r_h stand for home_result, where r_a stand for away_result
r_h,r_a = function_ha(football_df,teams)
```

### Q4. what is the ratio of scored goal per total matches  in each different Edition Fifa major(Continental) tournament?

```python
fifa_major_tournaments
tournament_df = football_df.copy()
tournament_df = tournament_df[(tournament_df.tournament).isin(fifa_major_tournaments)]
tournament_goals = tournament_df[tournament_df.tournament=="Copa América"]
```

```python
#collecte goals per edition
def goals_per_edition(tournament_goals):
  tournament_home_goals =  tournament_goals['home_score'].groupby(tournament_goals.date.dt.year).sum()
  tournament_away_goals  =  tournament_goals['away_score'].groupby(tournament_goals.date.dt.year).sum()
  total_goal = tournament_home_goals + tournament_away_goals
  total_goal
  goal_per_edition = pd.DataFrame(total_goal).sort_index(axis = 0) 
  goal_per_edition.columns = ['Total goals']
  goal_per_edition = goal_per_edition.reset_index()
  return  goal_per_edition
```

```python
#total of home matches played per team
def each_team_home_matches(tournament_goals):
  total_home_games_each_team = tournament_goals['home_team'].groupby(tournament_goals.date.dt.year).value_counts() 
  total_home_games_each_team
  each_team_home_games = pd.DataFrame(total_home_games_each_team).sort_index(axis = 0) 
  each_team_home_games.columns = ['Total home Games']
  each_team_home_games = each_team_home_games.reset_index()
  each_team_home_games = each_team_home_games.rename(columns={"home_team":"team"})
  return each_team_home_games
```

```python
# total of away matches played per team 
def each_team_away_macthes(tournament_goals):
  total_away_games_each_team = tournament_goals['away_team'].groupby(tournament_goals.date.dt.year).value_counts() #+ tournament_goals['away_team'].groupby(tournament_goals.date.dt.year).value_counts()
  total_away_games_each_team
  each_team_away_games = pd.DataFrame(total_away_games_each_team).sort_index(axis = 0) 
  each_team_away_games.columns = ['Total away Games']
  each_team_away_games = each_team_away_games.reset_index()
  each_team_away_games = each_team_away_games.rename(columns={"away_team":"team"})
  return each_team_away_games
```

```python
#create a dict of list containning edition number of goal and number of games
def edition_info(edition_years,each_team_home_games,each_team_away_games,goal_per_edition):
  total_games =[]
  total_goals = []
  edition = []
  for i in edition_years:
    # print(i)
    edition.append(i)
    total_games.append(each_team_home_games.loc[each_team_home_games["date"] == i, 'Total home Games'].sum())
    total_goals.append(goal_per_edition.loc[goal_per_edition.date==i,"Total goals"].iloc[0])
  return {"year":edition,
          "games":total_games,
          "goals":total_goals,}
```

```python
#lets do it in one function
def average_goals_per_edition(tournament_df):
  tournament_df = tournament_df[(tournament_df.tournament).isin(fifa_major_tournaments)]
  edition = []
  ed = {}
  j = 0
  for i in tournament_df.tournament.unique():
    tournament_goals = tournament_df[tournament_df.tournament==i]
    edition_years = tournament_goals.date.dt.year.unique()
    goals_per_ed = goals_per_edition(tournament_goals)
    each_team_home_games = each_team_home_matches(tournament_goals)
    each_team_away_games = each_team_away_macthes(tournament_goals)
    eachedition_goal_games_info = edition_info(edition_years, each_team_home_games, each_team_away_games, goals_per_ed)

    edition_game_goal = pd.DataFrame(eachedition_goal_games_info)
    edition_game_goal["average goals"] = edition_game_goal.goals / edition_game_goal.games
    edition.append(edition_game_goal)
    if i not in ed.keys():
      ed[i] = edition[j]
    j = j+1
  return ed
df= average_goals_per_edition(tournament_df)
```

```python
# plot the data
fig = go.Figure()

for i in df:
    fig = fig.add_trace(go.Scatter(x = df[i]["year"],
                                   y = df[i]["average goals"], 
                                   name = i))
    

fig.update_layout(title="Average Goals scored in each Continental Tournaments Edition",
                 xaxis_title = "Tournament Edition Year", yaxis_title= "Average Goals")
fig.show()
```

We can see That South America `Copa America` have record from `1916`, the smallest average goals was 2 in `1922` edition,where others cotinent have Edition with less than 2 goals average. This show how exciting `Copa America` is comparing to others where average scored goal are rarely more than 3. After `1980` beside `Ocenia cup` which has less challenging teams, others tournaments  have an average goals less than 3.3 it could also be due to the fact that nowaday many countries have done remarkable evolution in football development There is a real Challenge between teams now than before and matches are more tough too.
